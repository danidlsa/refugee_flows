## Country Borders

Got the data from [here](https://public.opendatasoft.com/explore/dataset/world-administrative-boundaries/export/)

```python
import country_converter as coco
import pandas as pd
coordinates = pd.read_json('world-administrative-boundaries.json').rename({'iso3':'country_code'}, axis=1)
cities = pd.read_csv('geonames-all-cities-with-a-population-1000.csv', sep=';', engine='pyarrow')
cities = cities[cities.Population > 200000]
cities['country_code'] = coco.convert(cities['Country Code'], to='iso3' )

# Function to convert a string to a tuple of coordinates
def convert_to_tuple(coordinates_str):
    latitude, longitude = coordinates_str.split(',')
    return (float(latitude), float(longitude))

# Apply the function to the series
cities['Coordinates'] = cities['Coordinates'].apply(convert_to_tuple)
```

```python
countries.reset_index(drop=True)[countries.reset_index(drop=True).str.contains( 'Canada')]
```

```python
from tqdm import tqdm
import json
from border_functions import *

bordering_country_cities = dict()
for country in tqdm(countries.tolist()):
    contigs = contiguous_countries(country)
    bordering_cities = dict()
    for contig in contigs:
        closest_cities = find_closest_cities(country,contig,coordinates, cities,2, full_names=True)
        if closest_cities is not None:
            bordering_cities.update({contig: closest_cities['ASCII Name'].to_list()})
    bordering_country_cities.update({country:bordering_cities})

with open('bordering_countries_bordering_cities.json', 'w') as fp:
    json.dump(bordering_country_cities, fp)
```
